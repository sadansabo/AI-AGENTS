```python
# 📌 GreenTime AI – Compost Checker Agent
# Google Colab Notebook

# ==============================
# 1. Introduction (Markdown cell in Colab)
# ==============================

"""
# ♻️ GreenTime AI – Compost Checker Agent

**Project Goal:**
Build an AI agent that classifies household items into **Compostable, Recyclable, or Trash**
using a Hugging Face transformer model.

**Why it matters:**
- 🌱 Helps reduce waste contamination
- ⚡ Simple, fast demo for sustainability projects
- 🌍 Encourages eco-friendly decision-making

---
"""

# ==============================
# 2. Install Dependencies
# ==============================
!pip install transformers torch --quiet

from transformers import pipeline

# ==============================
# 3. Load Hugging Face Model
# ==============================
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)

# ==============================
# 4. Define Compost Checker Function
# ==============================
def check_item(item: str):
    labels = ["compostable", "recyclable", "trash"]
    result = classifier(item, candidate_labels=labels)
    top = result["labels"][0]

    if top == "compostable":
        return "✅ Compostable"
    elif top == "recyclable":
        return "♻️ Recyclable"
    else:
        return "❌ Not Compostable (Trash)"

# ==============================
# 5. Quick Demo for Presentation
# ==============================
test_items = [
    "banana peel",
    "plastic bottle",
    "paper cup",
    "aluminum foil",
    "glass jar",
    "styrofoam plate"
]

print("♻️ GreenTime AI Compost Checker Demo:\n")
for item in test_items:
    print(f"Item: {item:<20} → {check_item(item)}")

# ==============================
# 6. Summary (Markdown cell in Colab)
# ==============================

"""
## ✅ Project Summary – Slide Ready

- 🌱 AI-powered waste classification assistant
- 🔍 Classifies items into **Compostable, Recyclable, or Trash**
- ⚡ Powered by Hugging Face transformer model (`facebook/bart-large-mnli`)
- 🧪 Tested in Google Colab with real examples
- 🌍 Can be deployed to Hugging Face Spaces or Streamlit

---
"""
```

```python
import gradio as gr
from transformers import pipeline

# Load a zero-shot classification model (online, Hugging Face Hub)
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Labels we want for waste classification
labels = ["Recyclable", "Compostable", "Trash"]

def classify_item(item):
    result = classifier(item, candidate_labels=labels)
    return dict(zip(result['labels'], [round(x, 3) for x in result['scores']]))

# Build Gradio UI
demo = gr.Interface(
    fn=classify_item,
    inputs=gr.Textbox(lines=2, placeholder="Enter a waste item (e.g., banana peel)"),
    outputs=gr.Label(num_top_classes=3),
    title="♻️ AI Waste Sorting Assistant",
    description="Enter a waste item, and the AI will classify it as Recyclable, Compostable, or Trash."
)

if __name__ == "__main__":
    demo.launch()
```

```python
# Install dependencies
!pip install -q transformers datasets accelerate

# Import libraries
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
import torch

# Load dataset (use AG News as example AI agent training data)
dataset = load_dataset("ag_news")

# Load tokenizer
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Tokenize
def tokenize(batch):
    return tokenizer(batch['text'], padding="max_length", truncation=True)

dataset = dataset.map(tokenize, batched=True)
dataset = dataset.rename_column("label", "labels")
dataset.set_format("torch", columns=['input_ids', 'attention_mask', 'labels'])

# Split
train_dataset = dataset["train"].shuffle(seed=42).select(range(20000))   # subset for speed
test_dataset  = dataset["test"].shuffle(seed=42).select(range(5000))

# Load model
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=4)

# Training arguments
training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=2,
    weight_decay=0.01,
    logging_dir='./logs',
)

# Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    tokenizer=tokenizer,
)

# Train
trainer.train()

# Evaluate
metrics = trainer.evaluate()
print(metrics)

# Save model
model.save_pretrained("ai_agent_model")
tokenizer.save_pretrained("ai_agent_model")
```

```python
results = trainer.evaluate()
print(results)

# Test prediction
sample = "Apple shares rise as company announces new iPhone"
inputs = tokenizer(sample, return_tensors="pt")
with torch.no_grad():
    logits = model(**inputs).logits
pred = torch.argmax(logits, dim=-1).item()
print("Predicted class:", pred)
```
